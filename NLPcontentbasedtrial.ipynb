```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
df2 = pd.read_csv('https://msi.martial.co.ke/data/BX-Books.csv')
df2.head(1000)
```

```python
#get the information
df2.info()
```

```python
#get the content - book title
content = df2['Book-Title'][0:20]
content
```

```python
!pip install sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('distilbert-base-nli-stsb-mean-tokens')
```

```python
embeddings = model.encode(content, show_progress_bar=True)
```

```python
embeddings_df = pd.DataFrame(embeddings)
embeddings_df.head(20)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cos_sim = cosine_similarity(embeddings_df)
cos_sim
```

```python
cosin_sim_df = pd.DataFrame(cos_sim)
cosin_sim_df.head(20)
```

```python
title_index = cosin_sim_df.loc[15].sort_values(ascending=False).index.to_list()[0:18]
title_index
```

```python
df2.head()
```

```python
book_title = df2['Book-Title'].loc[title_index]
book_title
```
